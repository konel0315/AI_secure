```python
from sklearn.preprocessing import LabelEncoder
# LabelEncoder 객체 생성
label_encoder = LabelEncoder()

df[' Source IP'] = label_encoder.fit_transform(df[' Source IP'])
df[' Destination IP'] = label_encoder.fit_transform(df[' Destination IP'])
df['Label'] = label_encoder.fit_transform(df['Label'])

# X_data는 Label 열을 제외한 데이터, Y_data는 Label 열
X_data = df.drop(columns=['Label'], axis=1)
Y_data = df['Label']

# inf, -inf 값을 0으로 대체
X_data.replace([np.inf, -np.inf], 0, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(X_data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_data,Y_data,test_size=0.20,random_state=2222)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# 개별 모델 정의
logistic_model = LogisticRegression(max_iter=300)
tree_model = DecisionTreeClassifier(random_state=42)
svm_model = SVC(probability=True)

# 소프트 투표 방식의 VotingClassifier 정의
voting_model_soft = VotingClassifier(
    estimators=[('logistic', logistic_model), 
                ('tree', tree_model), 
                ('svm', svm_model)], 
    voting='soft'
)

# 모델 학습
voting_model_soft.fit(x_train, y_train)
```

```python
# 테스트 데이터에 대해 예측
final_predictions_soft = voting_model_soft.predict(x_test)

# 필요한 성능 지표들을 임포트
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 정확도 계산
accuracy = accuracy_score(y_test, final_predictions_soft)
print(f"Accuracy: {accuracy:.4f}")

# 정밀도 계산
precision = precision_score(y_test, final_predictions_soft, average='macro')
print(f"Precision: {precision:.4f}")

# 재현율 계산
recall = recall_score(y_test, final_predictions_soft, average='macro')
print(f"Recall: {recall:.4f}")

# F1 점수 계산
f1 = f1_score(y_test, final_predictions_soft, average='macro')
print(f"F1 Score: {f1:.4f}")
```

```python
voting_model_hard = VotingClassifier(
    estimators=[('logistic', logistic_model), 
                ('tree', tree_model), 
                ('svm', svm_model)], 
    voting='hard'
)

voting_model_hard.fit(x_train, y_train)
```

```python
# 테스트 데이터에 대해 예측
final_predictions_hard = voting_model_hard.predict(x_test)

# 필요한 성능 지표들을 임포트
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 정확도 계산
accuracy = accuracy_score(y_test, final_predictions_hard)
print(f"Accuracy: {accuracy:.4f}")

# 정밀도 계산
precision = precision_score(y_test, final_predictions_hard, average='macro')
print(f"Precision: {precision:.4f}")

# 재현율 계산
recall = recall_score(y_test, final_predictions_hard, average='macro')
print(f"Recall: {recall:.4f}")

# F1 점수 계산
f1 = f1_score(y_test, final_predictions_hard, average='macro')
print(f"F1 Score: {f1:.4f}")
```
