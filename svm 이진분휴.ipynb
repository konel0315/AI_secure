```python
df['label'] = df['label'].map({'mitm': 1, 'other': 0}).fillna(0)
```

```python
df['label']
```

```python
train_data:int=df.drop(columns = [
    'src_ip','dst_ip','src_mac','dst_mac','timestamp','fwd_psh_flags','bwd_psh_flags', 
    'fwd_urg_flags','bwd_urg_flags','syn_flag_cnt','rst_flag_cnt','psh_flag_cnt','ack_flag_cnt','urg_flag_cnt','ece_flag_cnt','label'],axis=1)
label_data:int=df['label']
```

```python
label_data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data,label_data,test_size=0.20,random_state=2202)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(x_train)
X_test=scaler.transform(x_test)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear',C=100)
svm.fit(X_train,y_train)
```

```python
y_pred_test = svm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Accuracy 계산
accuracy = accuracy_score(y_test, y_pred_test)
print(f"Accuracy: {accuracy:.4f}")

# Precision 계산
precision = precision_score(y_test, y_pred_test)
print(f"Precision: {precision:.4f}")

# Recall 계산
recall = recall_score(y_test, y_pred_test)
print(f"Recall: {recall:.4f}")

# F1 Score 계산
f1 = f1_score(y_test, y_pred_test)
print(f"F1 Score: {f1:.4f}")
```
